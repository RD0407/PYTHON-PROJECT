```python
import numpy as np
import matplotlib.pyplot as plt
m=0.8
theta=np.arange(0,np.pi,0.01)
x=np.cos(theta)
s=7.56
beta=0.9091
gamma=1/(np.sqrt(1-(beta**2)))
P= (s-(m**2))/(2*np.sqrt(s))
E=np.sqrt(P**2 + m**2)
def f(x):
    return(np.arctan((P*(np.sqrt(1-(x*x))))/(gamma*((P*x)+(beta*E)))))
y=f(x)
#print(y)
plt.plot(theta, y)
```

```python
m1=1.2
theta=np.arange(0,np.pi,0.01)
x=np.cos(theta)
s=7.56
beta=0.9091
gamma=1/(np.sqrt(1-(beta**2)))
P1= (s-(m1**2))/(2*np.sqrt(s))
E1=np.sqrt(P1**2 + m1**2)
def f1(x):
    return(np.arctan((P1*(np.sqrt(1-(x*x))))/(gamma*((P1*x)+(beta*E1)))))
y1=f1(x)
plt.plot(np.degrees(theta), np.degrees(y))
plt.plot(theta, y1)
```
